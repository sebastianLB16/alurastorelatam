### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)
tienda.info()

tienda.head()
```

#1. Análisis de facturación



```python
ingreso_total_tienda1 = tienda["Precio"].sum()
ingreso_total_tienda2 = tienda2["Precio"].sum()
ingreso_total_tienda3 = tienda3["Precio"].sum()
ingreso_total_tienda4 = tienda4["Precio"].sum()
print(ingreso_total_tienda1)
print(ingreso_total_tienda2)
print(ingreso_total_tienda3)
print(ingreso_total_tienda4)
```

```python
import matplotlib.pyplot as plt

# Datos
ingresos = [ingreso_total_tienda1, ingreso_total_tienda2, ingreso_total_tienda3, ingreso_total_tienda4]
tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']

# Gráfico
plt.figure(figsize=(7,5))
plt.bar(tiendas, ingresos, color='royalblue')
plt.title('Ingresos totales por tienda')
plt.xlabel('Tienda')
plt.ylabel('Monto total ($)')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

# 2. Ventas por categoría

```python
ingreso_por_categoria_tienda1 = tienda.groupby("Categoría del Producto")["Precio"].sum().sort_values(ascending=False)
ingreso_por_categoria_tienda2 = tienda2.groupby("Categoría del Producto")["Precio"].sum().sort_values(ascending=False)
ingreso_por_categoria_tienda3 = tienda3.groupby("Categoría del Producto")["Precio"].sum().sort_values(ascending=False)
ingreso_por_categoria_tienda4 = tienda4.groupby("Categoría del Producto")["Precio"].sum().sort_values(ascending=False)
print("Tienda 1 \n" ,ingreso_por_categoria_tienda1,"\n tienda 2")
print(ingreso_por_categoria_tienda2,"\n tienda 3")
print(ingreso_por_categoria_tienda3,"\n tienda 4")
print(ingreso_por_categoria_tienda4,"\n ======================")
ingreso_por_categoria_tienda1.plot.bar()



```

```python

def graficar_pie_legible(ingreso_por_categoria_df, titulo_grafico='Distribucion de Ingresos por Categoria'):

    total_ingresos = ingreso_por_categoria_df.sum()
    umbral = total_ingresos * 0.02

    #Identificar y agrupar categorias pequenas
    categorias_pequenas = ingreso_por_categoria_df[ingreso_por_categoria_df < umbral]
    categorias_grandes = ingreso_por_categoria_df[ingreso_por_categoria_df >= umbral]

    #Crear la categoria 'Otros'
    if not categorias_pequenas.empty:
        otros_ingresos = categorias_pequenas.sum()
        #Crear la nueva Serie para graficar
        ingresos_ajustados = categorias_grandes.copy()
        ingresos_ajustados['Otros'] = otros_ingresos
        #Ordeno para que 'Otros' no quede necesariamente al final
        ingresos_ajustados = ingresos_ajustados.sort_values(ascending=False)
    else:
        ingresos_ajustados = ingreso_por_categoria_df.copy()

    #Generar el gráfico de pastel
    plt.figure(figsize=(8, 8))
    ingresos_ajustados.plot(
        kind='pie',
        autopct='%1.1f%%',
        startangle=90,
        title=titulo_grafico,
        wedgeprops={'edgecolor': 'black', 'linewidth': 0.5},
        textprops={'fontsize': 10}
    )
    plt.ylabel('')
    plt.show()

graficar_pie_legible(ingreso_por_categoria_tienda1, 'Distribucion de Ingresos por Categoria - Tienda 1')
```

```python
graficar_pie_legible(ingreso_por_categoria_tienda2, 'Distribucion de Ingresos por Categoria - Tienda 2')
```

```python
graficar_pie_legible(ingreso_por_categoria_tienda3, 'Distribucion de Ingresos por Categoria - Tienda 3')
```

```python
graficar_pie_legible(ingreso_por_categoria_tienda4)
```

# 3. Calificación promedio de la tienda


```python
calificacion_promedio_tienda1=tienda['Calificación'].mean()
calificacion_promedio_tienda2=tienda2['Calificación'].mean()
calificacion_promedio_tienda3=tienda3['Calificación'].mean()
calificacion_promedio_tienda4=tienda4['Calificación'].mean()
print(calificacion_promedio_tienda1)
print(calificacion_promedio_tienda2)
print(calificacion_promedio_tienda3)
print(calificacion_promedio_tienda4)
```

```python
calificaciones = [
    calificacion_promedio_tienda1,
    calificacion_promedio_tienda2,
    calificacion_promedio_tienda3,
    calificacion_promedio_tienda4
]
nombres = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']

plt.bar(nombres, calificaciones, color='orange')
plt.title('Valoración promedio de clientes por tienda')
plt.ylabel('Promedio de calificación')
plt.ylim(0, 5)
plt.show()
```

# 4. Productos más y menos vendidos

```python
tabla_de_productos_mas_vendidos_tienda1 = tienda.groupby("Producto")["Precio"].sum().sort_values(ascending=False)
tabla_de_productos_mas_vendidos_tienda2 = tienda2.groupby("Producto")["Precio"].sum().sort_values(ascending=False)
tabla_de_productos_mas_vendidos_tienda3 = tienda3.groupby("Producto")["Precio"].sum().sort_values(ascending=False)
tabla_de_productos_mas_vendidos_tienda4 = tienda4.groupby("Producto")["Precio"].sum().sort_values(ascending=False)
print("tienda numero 1 \ntabla general de productos mas vendidos",tabla_de_productos_mas_vendidos_tienda1)
print("tienda numero 2 \ntabla general de productos mas vendidos",tabla_de_productos_mas_vendidos_tienda2)
print("tienda numero 3 \ntabla general de productos mas vendidos",tabla_de_productos_mas_vendidos_tienda3)
print("tienda numero 4 \ntabla general de productos mas vendidos",tabla_de_productos_mas_vendidos_tienda4)
def graficar_ventas_por_producto(tabla_ventas, titulo_grafica="Productos Más Vendidos", top_n=10):

    datos_a_graficar = tabla_ventas.head(top_n)
    plt.figure(figsize=(10, 6))
    datos_a_graficar.plot.bar(color='teal', edgecolor='black')

    plt.title(titulo_grafica, fontsize=15, fontweight='bold')
    plt.xlabel('Producto', fontsize=12)
    plt.ylabel('Ventas Totales (Suma de Precio)', fontsize=12)

    # 5. Formatear el eje X y ajustar el layout
    plt.xticks(rotation=45, ha='right') # Rotar etiquetas para mejor lectura
    plt.grid(axis='y', linestyle='--', alpha=0.7) # Añadir líneas de cuadrícula
    plt.tight_layout() # Asegura que nada se corte

    # 6. Mostrar la gráfica
    plt.show()
```

```python
graficar_ventas_por_producto(tabla_de_productos_mas_vendidos_tienda1)
```

```python
graficar_ventas_por_producto(tabla_de_productos_mas_vendidos_tienda2)
```

```python
graficar_ventas_por_producto(tabla_de_productos_mas_vendidos_tienda3)
```

```python
graficar_ventas_por_producto(tabla_de_productos_mas_vendidos_tienda4)
```
